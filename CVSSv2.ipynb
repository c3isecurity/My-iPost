# CVSS Temporal Score Function

```python
# CVSS Temporal Score
# Temporal E:[U,POC,F,H,ND]/RL:[OF,TF,W,U,ND]/RC:[UC,UR,C,ND]
# E - Exploitability
# RL - Remediation Level
# RC - Report Confidence

def cvss_temporal(BaseScore, E, RL, RC):
# Exploitability 
    if E == "U":
        Exploitability = .85
    elif E == "POC":
        Exploitability = .9
    elif E == "F":
        Exploitability = .95
    elif E == "H":
        Exploitability = 1
    elif E == "ND":
        Exploitability = 1
    
# Remediation Level
    if RL == "OF":
        RemedationLevel = .87
    elif RL == "TF":
        RemedationLevel = .9
    elif RL == "W":
        RemedationLevel = .95
    elif RL == "U":
        RemedationLevel = 1
    elif RL == "ND":
        RemedationLevel = 1

# Report Confidence
    if RC == "UC":
        ReportConfidence = .9
    elif RC == "UR":
        ReportConfidence = .95
    elif RC == "C":
        ReportConfidence = 1
    elif RC == "ND":
        ReportConfidence = 1
        

# Temporal Equation
    TemporalScore = BaseScore * Exploitability * RemedationLevel * ReportConfidence
    
    #print TemporalScore
    return TemporalScore
    return ReportConfidence
    
cvss_temporal(9, "U", "OF", "C")
#print ReportConfidence
```
